```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/Loan-Approval-Prediction.csv")
data.head()
```

```python
data.drop(["Loan_ID"],axis = 1,inplace = True)
```

```python
data.isnull().sum()
```

```python
obj = (data.dtypes == 'object');
object_col = list(obj[obj].index);
plt.figure(figsize = (18,36));
index =1;
for col in object_col:
    y = data[col].value_counts();
    plt.subplot(11,4,index);
    plt.xticks(rotation=90);
    sns.barplot(x=list(y.index), y=y); 
    index +=1;
```

```python
# Import label encoder 
from sklearn import preprocessing 
	
# label_encoder object knows how 
# to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
obj = (data.dtypes == 'object') 
for col in list(obj[obj].index): 
    data[col] = label_encoder.fit_transform(data[col])
```

```python
print("Categorical variables:",len(list(obj[obj].index)))
```

```python
plt.figure(figsize=(12,6)) 
  
sns.heatmap(data.corr(),cmap='BrBG',fmt='.2f', 
            linewidths=2,annot=True)
```

```python
for col in data.columns:
    data[col] = data[col].fillna(data[col].mean())
data.isnull().sum()
```

```python
X = data.drop(['Loan_Status'],axis = 1)
Y = data['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.35 , random_state =3)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_1 = DecisionTreeClassifier(max_depth = 25,max_features = 6,max_leaf_nodes = 15)
model_1.fit(X_train,Y_train)
print("Train accuracy" + str(model_1.score(X_train,Y_train)))
print("test accuracy" + str(model_1.score(X_test,Y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_2 = RandomForestClassifier(n_estimators = 20, 
                             max_leaf_nodes = 12,random_state = 42,
                             max_depth  =50)
model_2.fit(X_train,Y_train)
print("Train accuracy" + str(model_2.score(X_train,Y_train)))
print("test accuracy" + str(model_2.score(X_test,Y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
model_3 = LogisticRegression(max_iter = 200);
model_3.fit(X_train,Y_train);
print("Train accuracy" + str(model_3.score(X_train,Y_train)));
print("test accuracy" + str(model_3.score(X_test,Y_test)));
```

```python
y_pred = model_3.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
```

```python
plt.figure(figsize=(10, 7))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
# tick_marks = np.arange(len(set(Y_test)))
# plt.xticks(tick_marks, tick_marks)
# plt.yticks(tick_marks, tick_marks)

# Labeling the plot
plt.ylabel('True label')
plt.xlabel('Predicted label')

# Add values inside the matrix
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], 'd'),
             ha="center", va="center",
             color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```
